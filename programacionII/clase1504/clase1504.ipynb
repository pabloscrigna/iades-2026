List Comprehensions

Supongamos que tenemos una lista de numeros y queremos elevar esos numeros al cuadrado

```python
numeros = list(range(10))
numeros
```

```python
dobles = []

for n in numeros:
    dobles.append(n**2)

dobles
```

¿Hay otra forma de hacerlo?

```python
foo = [ n**2 for n in numeros]
foo
```

List Comprehensions

- expresión
- Iteracion

- Primero se ejecuta la iteración y después la expresión.
- El resultado de una list comprehension es una lista
- La nueva lista es el resultado de evaluar la expresión para cada uno de los elementos de mi lista.
- la nueva lista tiene la misma cantidad de elementos que la lista original.


```python
[
    n**2                        # expresión
    for n in numeros            # iteración
]
```

Hacer una lista de diccionarios [{ marca, color, talle}, {marca, color talle}]

```python
"    hola  mundo !!!! ".strip()
```

```python
# Adidas	orange	43
def linea_a_diccionario(linea: str) -> dict:
  campos = linea.strip().split('\t')
  return {
      "marca": campos[0],
      "color": campos[1],
      "talle": campos[2]
  }
```

```python
print("New Balance\torange\t78")
```

```python
"New Balance	orange	78".split("\t")
```

```python
linea_a_diccionario("Adidas	orange	43")
```

```python
[
    linea_a_diccionario(linea)                                               # expresion
    for linea in open("/content/drive/MyDrive/samples/shoe-data.txt")        # iteración
]
```

Dict Comprehension

*dado un string,  crer un diccionario, donde cada palabra sea la key y el valor la longitud de la palabra*

{
  "estamos": 5,
  "aprendiendo": 11,
}

```python
frase = "estamos aprendiendo a programar en python"
```

```python
dict([("estamos", 7), ("aprendiendo", 11)])
```

```python
frase = "estamos aprendiendo a programar en python"

diccionario = {}
diccionario_2 = {}

for palabra in frase.split():
  diccionario[palabra] = len(palabra)
  diccionario_2 = { **diccionario_2, palabra: len(palabra) }

print(diccionario)
print(diccionario_2)
```

```python
hh="estamos aprendiendo a programar en python"

palabras = hh.split()

dicionario = {
    palabra : len(palabra)
    for palabra in palabras
}

print(dicionario)
```

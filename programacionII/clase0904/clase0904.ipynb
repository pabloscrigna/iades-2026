Lo mismo que el ejemplo anterior, con una clase

```python
# las clases tienen: atributos y metodos
class Multiplicador:

  # factor : es un atributo de instancia
  def __init__(self, factor):
    self.factor = factor

  def __str__(self):
    return f"Factor: {self.factor}"

  # método
  def multiplicador(self, x):
    return x * self.factor
```

```python
# es una instancia de la clase Multiplicador -atributo factor = 5
multi_5 = Multiplicador(5)
```

```python
multi_5.multiplicador(20)
```

```python
print(multi_5)
```

```python
class Multiplicador:

  def __init__(self, factor):
    self.factor = factor

  def __call__(self, x):
    return x * self.factor
```

```python
# multi_5 es una instancia de Multiplicador con el atributo factor igual a 5
multi_5 = Multiplicador(5)
```

```python
multi_5(20)
```

¿Qué es un decorador?

```python
import time

def cuadrado(n):

  # delay de n segundos
  time.sleep(n)

  return n ** 2
```

```python
import time

from functools import lru_cache

@lru_cache
def cuadrado(n):

  # delay de n segundos
  time.sleep(n)

  return n ** 2
```

```python
cuadrado(10)
```

```python
cuadrado(8)
```

Limitaciones:

- No podemos pasarle argumentos
- No podemos retornar valores a traves del decorador

Vamos a ver como solucionamos estas limitaciones...

Pasarle argumentos:

```python
def wrapper(func):

  def _wrapper(name):
    print(f"antes de llamar a la funcion: {func.__name__}")
    func(name)
    print(f"despues de llamar a la funcion: {func.__name__}")

  return _wrapper
```

```python
@wrapper
def saludar(nombre):
  print(f"Hola {nombre}")
```

```python
saludar("Martin")
```

```python
def wrapper(func):

  def _wrapper(*args, **kwargs):
    print(f"antes de llamar a la funcion: {func.__name__}")
    func(*args, **kwargs)
    print(f"despues de llamar a la funcion: {func.__name__}")

  return _wrapper
```

```python
@wrapper
def saluda_compuesto(nombre, apellido):
  print(f"Hola {nombre} {apellido}")
  return
```

```python
saluda_compuesto("Juan", "Martin")
```

```python
@wrapper
def saludar(nombre):
  print(f"Hola {nombre}")
```

```python
saludar("Martin")
```

Retorne valores

```python
def wrapper(func):

  def _wrapper(*args, **kwargs):
    print(f"antes de llamar a la funcion: {func.__name__}")
    valor = func(*args, **kwargs)
    print(f"despues de llamar a la funcion: {func.__name__}")
    return valor

  return _wrapper
```

```python
import random

@wrapper
def aleatorio():
  return random.randint(1, 10)
```

```python
aleatorio()
```

Hacer un decorador que mida el tiempo que tarda en ejecutarse una funcion

```python
import time

def mide_tiempo(func):

  def _wrapper(*args, **kwargs):
    # iniciamos un timer
    cuenta_inicial = time.perf_counter()
    valor = func(*args, **kwargs)
    # terminamos timer
    cuenta_final = time.perf_counter()
    tiempo_transcurrido = cuenta_final - cuenta_inicial
    print(f"Tiempo transcurrido: {tiempo_transcurrido}")
    return valor

  return _wrapper
```

```python
import time

@mide_tiempo
def calcular_demora(valor):
  time.sleep(valor)
  return
```

```python
@mide_tiempo
def saludar(nombre):
  print(f"Hola {nombre}")
```
